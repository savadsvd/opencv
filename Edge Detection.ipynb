### robert operator

```python
import cv2
```

```python
#roberts kernal
import numpy as np
import matplotlib.pyplot as plt
kernalx=np.array([[-1,0],[0,1]],dtype=int)
kernaly=np.array([[0,-1],[1,0]],dtype=int)
```

```python
img=cv2.imread(r'C:\Users\Lenovo\Downloads\flower.jpg',0)
#cv2.filter2D(img,depth of destination image,kernal)
x=cv2.filter2D(img,cv2.CV_16S,kernalx)   #16 bit signed integers
y=cv2.filter2D(img,cv2.CV_16S,kernaly)
```

```python
# turn uint8,image fusion
#absolute  values are calculated and image converted to 8-bit
absX=cv2.convertScaleAbs(x)
absY=cv2.convertScaleAbs(y)
Roberts=cv2.addWeighted(absX,0.5,absY,0.5,0)
```

```python
plt.imshow(absX,'gray')
```

### prewitt operator

```python
kernalx=np.array([[1,0,-1],[1,0,-1],[1,0,-1]],dtype=int)
kernaly=np.array([[1,1,1,],[0,0,0],[-1,-1,-1]],dtype=int)
```

```python
img=cv2.imread(r'C:\Users\Lenovo\Downloads\flower.jpg',0)
x=cv2.filter2D(img,cv2.CV_16S,kernalx)   
y=cv2.filter2D(img,cv2.CV_16S,kernaly)
absX=cv2.convertScaleAbs(x)
absY=cv2.convertScaleAbs(y)
prewitt=cv2.addWeighted(absX,0.5,absY,0.5,0)
```

```python
plt.imshow(prewitt,'gray')
```

### sobel operator

```python
#cv2.CV_64F-----depth odf destination image same as input image
#1--order of the derivative of x
#0--order of the derivative of y
sobelx=cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)   #64 bit floating point nmbers
sobely=cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
i=cv2.addWeighted(sobelx,0.5,sobely,0.5,0)
```

```python
cv2.imshow('vertical_detection',sobelx)
cv2.imshow('horizontel_detection',sobely)
cv2.imshow('sobel_image',i)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### laplace  operator

```python
laplace=cv2.Laplacian(img,cv2.CV_64F,ksize=7)
cv2.imshow('laplace',laplace)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
